```python
import pandas as pd
```

```python
# Define the file paths for the Excel files
file_paths = ['./getting_data/sp_data_10y.xlsx', './getting_data/nas_data_10y.xlsx',
              './getting_data/vix_data_10y.xlsx', './getting_data/rf_data_10y.xlsx',
              './getting_data/shci_data_10y.xlsx']

# Define the column names to extract from each file
column_names = ['return', 'return', 'return', 'return', 'return']

# Create an empty DataFrame to store the extracted data
combined_data = pd.DataFrame()

# Read and extract the specified column from each file
for file_path, column_name in zip(file_paths, column_names):
    df = pd.read_excel(file_path)  # Read the Excel file
    df = df[['Date', column_name]]  # Extract the specified column
    df = df.rename(columns={column_name: column_name + '_' + file_path})  # Rename the column
    if combined_data.empty:
        combined_data = df  # Assign the first DataFrame to combined_data
    else:
        combined_data = pd.merge(combined_data, df, on='Date', how='inner')  # Merge the DataFrame based on the common date
```
