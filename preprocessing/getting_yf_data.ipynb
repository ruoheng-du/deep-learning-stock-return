```python
start_date = '2003-06-01'
end_date = '2023-06-01'
```

```python
# S&P 500
sp = '^GSPC'
sp_data = yf.download(sp, start = start_date, end = end_date, interval='1d')
```

```python
# sp return
sp_data.sort_values('Date', inplace=True)
sp_data['return'] = (sp_data['Close'] - sp_data['Close'].shift(1)) / sp_data['Close'].shift(1)

# 删除首行
sp_data.dropna(subset=['return'], inplace=True)
```

```python
sp_data.to_excel('sp_data_20y.xlsx')
```
